#**LISTAS EN PYTHON**

**1. El Inventario:**

Enunciado: Crea una lista llamada frutas con "manzana", "banana" y "cereza". Luego, añade "naranja" al final y cambia "banana" por "pera".

```python
frutas = ["manzana", "banana", "cereza"]
frutas.append("naranja")
#frutas[1] = "pera" con esta se elimina la de la posición uno y pone la nueva, en la otra función solo introduce
frutas.insert(1, "pera")
print(frutas)
```

**2. Suma de una lista**

Enunciado: Tienes la lista de precios [10, 5, 20]. Usa la función sum() de Python para mostrar el total de la cuenta.

```python
numeros = [10, 5, 20] #creo una cadena con todos los números y con la función los separa y los suma
suma = sum(numeros)
print(suma)
print(sum(numeros)) #es otra manera de sumar directamente pero es mejor en una variable como la anterior
```

**3. Acceso Inverso:**

Dada la lista colores = ["rojo", "azul", "verde", "amarillo"], imprime el último elemento usando un índice negativo.

```python
colores = ["rojo", "azul", "verde", "amarillo"]
print(colores[-1])
```

**4. ¿Cuántos hay?**

Enunciado: Dada la lista notas = [10, 7, 8, 10, 5, 10], cuenta cuántas veces aparece la nota 10.*italicized text*

```python
notas = [10, 7, 8, 10, 5, 10]
cantidad = notas.count(10)
print(cantidad)
print(f "La nota 10 aparece {cantidad} veces")
```

**4. El primero y el último**

Enunciado: Dada la lista frutas = ["manzana", "pera", "platano", "kiwi"], imprime solo la primera y la última fruta (usa índices).

```python
frutas = ["manzana", "pera", "platano", "kiwi"]
primera = frutas[0]
ultima = frutas[-1]
print(primera, ultima)
```

**6. Ordenando la fila**

Enunciado: Tienes una lista de números desordenados [5, 2, 9, 1]. Ordénalos de menor a mayor usando el método .sort().

```python
numeros = [5, 2, 9, 1]
numeros.sort(reverse=True)
print(numeros)
```

**Buscador de Números**

**Enunciado:** Dada la lista numeros = [10, 20, 30, 40, 50, 20, 60].

1. Encuentra en qué índice se encuentra el número 30.

2. Cuenta cuántas veces aparece el número 20.

3. Invierte el orden de la lista completa.

```python
numeros = [10, 20, 30, 40, 50, 20, 60]
indice = numeros.index(30)
repetidos = numeros.count(20)
numeros.sort(reverse=True)

print(f"el numero 30 tiene un índice de {indice}.")
print(f"¿Cuántas veces se repite el 20?: {repetidos}")
print(f"la lista inversa es: {numeros}")
```

**El Gestor de Invitados**

**Enunciado:** Tienes una lista inicial de invitados: invitados = ["Ana", "Juan", "Pedro"].

* Añade a "Maria" al final de la lista.

* Añade a "Luis" en la segunda posición (índice 1).

* Elimina a "Juan" de la lista usando su nombre.

* Imprime cuántos invitados hay en total usando una función.

```python
invitados = ["Ana", "Juan", "Pedro"]
invitados.append("Maria")
invitados.insert(1,"Luis")
invitados.remove("Juan")
total_invitados = len(invitados)
print(f"La nueva lista de invitados es: {invitados}, y hay un total de {total_invitados} invitados.")
```
